# Regularization Technique

```python
data_c = pd.read_csv("diabetes.csv")
```

```python
data_c.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=data_c.corr() , annot=True)
plt.show()
```

```python
x = data_c.iloc[:,:-1]
y = data_c["Outcome"]
```

```python
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x) , columns=x.columns)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LinearRegression , Lasso , Ridge
from sklearn.metrics import mean_absolute_error , mean_squared_error
```

Linear regression

```python
lr = LinearRegression()
lr.fit(x_train , y_train)
lr.score(x_test , y_test)*100
```

```python
# for Linear Regression
print(mean_squared_error(y_test , lr.predict(x_test))) # MSE
print(mean_absolute_error(y_test , lr.predict(x_test))) # MAE
print(np.sqrt(mean_squared_error(y_test , lr.predict(x_test)))) #RMSE

```

```python
plt.figure(figsize=(15,6))
plt.title("Linear Regression")
plt.bar(x.columns , lr.coef_)
plt.xlabel("Linear Regression")
plt.ylabel("Coef")
plt.show()
```

# Lasso 

```python
la = Lasso(alpha=1)
la.fit(x_train , y_train)
la.score(x_test , y_test)*100
```

```python
# for Lasso
print(mean_squared_error(y_test , la.predict(x_test))) # MSE
print(mean_absolute_error(y_test , la.predict(x_test))) # MAE
print(np.sqrt(mean_squared_error(y_test , la.predict(x_test)))) #RMSE

```

```python
plt.figure(figsize=(15,6))
plt.title("Lasso")
plt.bar(x.columns , la.coef_)
plt.xlabel("Lasso")
plt.ylabel("Coef")
plt.show()
```

# Ridge

```python
ri = Ridge(alpha=1)
ri.fit(x_train , y_train)
ri.score(x_test , y_test)*100
```

```python
# for Ridge
print(mean_squared_error(y_test , ri.predict(x_test))) # MSE
print(mean_absolute_error(y_test , ri.predict(x_test))) # MAE
print(np.sqrt(mean_squared_error(y_test , ri.predict(x_test)))) #RMSE

```

```python
plt.figure(figsize=(15,6))
plt.title("Ridge")
plt.bar(x.columns , ri.coef_)
plt.xlabel("Ridge")
plt.ylabel("Coef")
plt.show()
```
