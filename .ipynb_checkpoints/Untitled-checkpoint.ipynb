# Ordinal Encoding

```python
df = pd.DataFrame({"Size": ["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df.head()
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
# BY using scikit learn
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
df["Size_en"] = oe.transform(df[["Size"]])
```

```python
df
```

```python
#By using map function:
ordered_data = {"s" :0 , "m":1 ,"l":2,"xl":3 }
```

```python
df["Size_en_map"] = df["Size"].map(ordered_data)
```

```python
df
```

# Outlier

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x="CoapplicantIncome", data=data)
plt.show()
```

```python
sns.distplot(data[["CoapplicantIncome"]])
plt.show()
```

```python
 data.shape
```

```python
# Outlier removal by IQR
q1 = data["CoapplicantIncome"].quantile(0.25)
q3 = data["CoapplicantIncome"].quantile(0.75)

```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5*IQR)
max_range = q3 + (1.5*IQR)
min_range , max_range
```

```python
# to remove outlier from the data
new_data = data[data["CoapplicantIncome"]<=max_range]
```

```python
new_data.shape
```

```python
sns.boxplot(x="CoapplicantIncome" , data = new_data)
```

```python
# outlier using Z score
out_data = pd.read_csv("loan.csv")
```

```python
out_data.isnull().sum()
```

```python
out_data.describe()
```

```python
sns.boxplot(x = "CoapplicantIncome" ,data = out_data)
```

```python
sns.distplot(out_data["CoapplicantIncome"])
```

```python
# find min and max range 
min_range = data["CoapplicantIncome"].mean() - (3 * data["CoapplicantIncome"].std())
max_range = data["CoapplicantIncome"].mean() + (3 * data["CoapplicantIncome"].std())
```

```python
min_range , max_range
```

```python
new_data = data[data["CoapplicantIncome"]<= max_range]
```

```python
sns.boxplot(x = "CoapplicantIncome" , data = new_data)
```

```python
z_score = (data["CoapplicantIncome"] - data["CoapplicantIncome"].mean()) /data["CoapplicantIncome"].std()
```

```python
z_score
```

```python
data["z_score"] = z_score
```

```python
 data[data["z_score"]<3]  
```

# Function Transformer

```python
sns.distplot(data["CoapplicantIncome"])
plt.show()
```

```python
q1 = data["CoapplicantIncome"].quantile(0.25)
q3 = data["CoapplicantIncome"].quantile(0.75)
iqr = q3 - q1
```

```python
min_r = q1-(1.5*iqr)
max_r = q3 + (1.5*iqr)
```

```python
data=data[data["CoapplicantIncome"]<=max_r]
```

```python
sns.distplot(data["CoapplicantIncome"])
plt.show()
# this shows a non normal distribution curve
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p)
```

```python
ft.fit(data[["CoapplicantIncome"]])
```

```python
data["CoapplicantIncome_tf"] = ft.transform(data[["CoapplicantIncome"]])
```

```python
plt.figure(figsize = (10,7))
plt.subplot(1,2,1)
sns.distplot(data["CoapplicantIncome"])
plt.title("Before")
plt.subplot(1,2,2)
sns.distplot(data["CoapplicantIncome_tf"])
plt.title("After")
```
